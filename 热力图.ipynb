```python

import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('AutoInsurance_v20221028 22-35.csv')

for i in ['State','Coverage','Education','EmploymentStatus','Location Code','Marital Status','Policy Type','Policy','Renew Offer Type','Sales Channel','Vehicle Class','Vehicle Size']:
    dummies = pd.get_dummies(df[i],prefix=str(i))
    df=df.join(dummies)
    df=df.drop(i, axis=1)

df['Response']=df['Response'].apply(lambda x : 1 if x=='Yes' else 0)
df['Gender']=df['Gender'].apply(lambda x: 1 if x=="M" else 0)
```

```python
df.to_csv('xxx.csv')
```

```python
features = df.columns
plt.figure(figsize=[12,10])
plt.title('Features Correlation-Plot')
sns.heatmap(df[features].corr(), vmin=-1, vmax=1, center=0) #, 
plt.show()
```
